# Pie Chart
A pie chart is a circle split into pieces like a pie. Each piece shows a different group. It's often used to show percentages, with all pieces adding up to 100%. The size of each piece depends on how many are in each group. A bigger piece means a more popular group. The number of items in a group is called the frequency.

Let's take an example and creates the pie chart on that.

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generating the dataset
df1 = pd.DataFrame({
    'Genre': ['Sports', 'Strategy', 'Action', 'Shooter', 'Other'],
    'UnitsSold': [27500, 11500, 6000, 3500, 1500]
})

df1
```

```python
# Setting up some customizations
plt.figure(figsize=(5, 5)) # Setting the figure size
sns.set(style='whitegrid') # Setting the background theme

# Creating a color palette
palette = sns.color_palette('pastel')[0:5]  # Using Seaborn color palette
```

```python
# Creating the pie chart
ax1 = plt.pie(
    df1['UnitsSold'],
    labels=df1['Genre'],
    autopct='%.0f%%',
    colors=palette,
    startangle = 90,
    textprops={'fontsize':9}
    )

# Adding a title
plt.title('Units Sold by Genre\n')

# Show the plot
plt.show()
```

**Interpretation**
*   `Sports` slice is much larger than all the others, which means that the frequency is highest for this category.
*   `Other` slice here is much smaller than the others, so this means sales are a lot lower for this category.

**How to explode the slice?**
\
Let's suppose we want to explode `Action` slice in the above pie chart.

```python
# Using 0.05 for exploding the slice for Action
myexplode = [0, 0, 0.05, 0, 0]

# Creating the pie chart
ax2 = plt.pie(
    df1['UnitsSold'],
    labels=df1['Genre'],
    autopct='%.0f%%',
    colors=palette,
    startangle = 90,
    textprops={'fontsize':9},
    explode = myexplode
    )

# Adding a title
plt.title('Units Sold by Genre\n')

# Show the plot
plt.show()
```

**So when are pie charts useful?**
\
Pie charts are good for comparing basic sizes of different groups. It's easy to see which groups are bigger. However, if all the pieces are similar in size, it can be hard to tell them apart. Also, we shouldn't plot pie charts when there are a lot of slices.

Let's take an example where we have the approx similar sectors.

```python
# Generating the dataset
df_new = pd.DataFrame({
    'Genre': ['Sports', 'Strategy', 'Action', 'Shooter', 'Other'],
    'UnitsSold': [1000, 1010, 1030, 1040, 1080]
})

df_new
```

```python
# Creating the pie chart
plt.pie(
    df_new['UnitsSold'],
    labels=df_new['Genre'],
    colors=palette
    )

# Adding a title
plt.title('Units Sold by Genre\n')

# Show the plot
plt.show()
```

Now it's difficult to tell which `Genre` have highest units solds.
