# Venn Diagram
A Venn Diagram (also called set diagram or logic diagram) shows the relationships between different sets. You can easily create one in Python using the **matplotlib_venn** library.
\
\
Each set in a Venn Diagram is represented by a circle. Sometimes, the circle size shows the importance of the group, but not always. The circles usually overlap, with the size of the overlap representing the intersection (common part) between the groups.
\
\
Let's understands this with an example:

```python
# Importing libraries
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn3 # Library to plot 2-set and 3-set Venn diagram
import seaborn as sns
import pandas as pd
```

```python
# Creating a dataframe
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [4, 5, 6, 7, 8]
})

df
```

```python
# Setting up some customizations
plt.figure(figsize=(8, 6)) # Setting the figure size
```

```python
# Convert the columns to sets
set_a = set(df['A'])
set_b = set(df['B'])

# Create a Venn diagram
ax1 = venn2(
    [set_a, set_b],
    ('A', 'B'),
    set_colors=('purple', 'orange')
    )

# Show the plot
plt.show()
```

In the above Venn diagram, we can observe that only 2 items are common between both groups. Let's plot one more Venn diagram with three groups.

```python
# Creating a new dataframe
df_new = pd.DataFrame({
    'C': [1, 3, 5, 7, 9]
})

df = pd.concat(
    [df, df_new],
    axis=1
)

df
```

```python
# Convert the columns to sets
set_c = set(df['C'])

# Create a Venn diagram
ax2 = venn3(
    [set_a, set_b, set_c],
    ('A', 'B', 'C'),
    set_colors=('purple', 'orange', 'lightsteelblue')
    )

# Show the plot
plt.show()
```

Venn diagrams are excellent for visualizing intersections between 2 or 3 sets. For more sets, they tend to become messy and hard to read.
