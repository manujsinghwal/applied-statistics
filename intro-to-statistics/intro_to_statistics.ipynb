# Statistics
Statistics are numbers that summarize raw data and converts it in some
meaningful way. It presents key ideas that may not be immediately
apparent by just looking at the raw data, and by data, we mean facts or figures from which we can draw conclusions.

### Why we need to learn statistics?
By understanding whats going on with data empowers you. If you understands statistics, you can make correct decisions and make accurate predictions and convey your message in most effective way. Statistics can be a convenient way of summarizing key truths about data, but thereâ€™s a dark side too.

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generating the dataset
df = pd.DataFrame({
    'Month': ['Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Profit (Millions)': ['2.0', '2.1', '2.2', '2.1', '2.3', '2.4']
})

df['Profit (Millions)'] = df['Profit (Millions)'].astype(float)
df.info()
```

```python
# Ploting the first line graph
sns.lineplot(data=df, x='Month', y='Profit (Millions)')

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Profit (Millions)')
plt.title('Monthly Profit (Millions)')

# Making some customizations for y-axis
plt.ylim(0, 2.5)
plt.yticks([0, 0.5, 1.0, 1.5, 2.0, 2.5])

# Showing the plot
plt.show()
```

By looking at the above graph, we can understand that there is not too much growth in profits.

```python
# Ploting the second line graph
sns.lineplot(data=df, x='Month', y='Profit (Millions)')

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Profit (Millions)')
plt.title('Monthly Profit (Millions)')

# Showing the plot
plt.show()
```

Now look at this graph, we can measure the significant diffrence in profits.

\
#### What's the difference?
Both charts are showing same information, but they convey the different message. The second chart gives a different impression by making the vertical axis start at a different place (2 instead of 0) and adjusting the scale accordingly. At a glance, the profits appear to be rising dramatically each month.
